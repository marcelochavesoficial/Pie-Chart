# **Pie Chart**

The pie chart is a circular diagram where the value of each class is proportional to its respective angle measure. In practice, this means that for every 1% of the total absolute value, there are 3.6 degrees on the pie.

Many people love putting pie charts on every report or survey. However, there are several reasons why you should avoid using this type of graphing.
The reason for this is that pie charts force us to compare areas or angles, which is not intuitive for the human brain. The situation gets worse if you want a person to compare two different pie charts.

# Plot the pie chart

The data that I will use in this notebook are public data from CENIPA - Aeronautical Occurrences in Brazilian Civil Aviation. The dataset containing these statistics can be found on the Brazilian Open Data Portal. The dataset used here, in csv format, can be downloaded from this link: https://bit.ly/3ZknaDv.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt

# import the csv file
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv",
                 index_col='codigo_ocorrencia')

# see the first entries
df.head()
```

I will identify the classification of occurrences in Brazilian civil aviation.

As the pie chart needs the total values ​​and labels, I will extract this information from the DataFrame for variables.

```python
# extract occurrences to a dictionary
ocorrencia_classificacao = dict(df.ocorrencia_classificacao.value_counts())

# prepare data for the pie chart
labels = ocorrencia_classificacao.keys()
sizes = ocorrencia_classificacao.values()
explode = (0.1, 0.1, 0.1)

```

```python
# plot the pie chart
fig, ax = plt.subplots(figsize=(6,6))

ax.pie(sizes, labels=labels)
ax.set_title("Classificação das Ocorrências")

plt.tight_layout()
```

```python
# plot the pie chart
fig, ax = plt.subplots(figsize=(6,6))

ax.pie(sizes, explode=explode, labels=labels, shadow=True )
ax.set_title("Classificação das Ocorrencias")

plt.tight_layout()
```

# Donut

```python
# creating the circle
donut = plt.Circle((0, 0), 0.70, fc='white')

# plot the pie chart
fig, ax = plt.subplots(figsize=(6,6))

ax.pie(sizes, labels=labels)
ax.set_title("Classificação das Ocorrências")
ax.add_artist(donut)

plt.tight_layout()
```
